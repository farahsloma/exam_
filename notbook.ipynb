```python
import pandas as pd
from langchain_core.documents import Document

df = pd.read_csv(r"C:\Users\Farah\Downloads\expr\data\All_Diets.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
docs = [
    Document(
        page_content=f"""
        Diet: {row['Diet_type']}
        Recipe: {row['Recipe_name']}
        Protein: {row['Protein(g)']}
        Carbs: {row['Carbs(g)']}
        Fat: {row['Fat(g)']}
        """
    )
    for _, row in df.iterrows()
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_split = RecursiveCharacterTextSplitter(chunk_size = 500 , chunk_overlap = 20)
splite = text_split.split_documents(docs) 
```

```python
splite
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
embedding = HuggingFaceEmbeddings(model='sentence-transformers/all-MiniLM-L6-v2')
```

```python
from langchain_chroma import Chroma
```

```python
vdb = Chroma.from_documents(
    embedding = embedding,
    documents = splite,
    persist_directory = 'diet_food'
)
```
